```python
containers = [x.get_attribute('innerHTML') for x in browser.find_elements(By.XPATH,"//div[@class='review-container']")]
len(containers)
```

```python
import re
```

```python
reviews = []
ratings=[]
for container in containers:
  bsObj = BeautifulSoup(container)
  rev = bsObj.find(name="div",attrs = {'class':'text show-more__control'}).text
  reviews.append(rev)

  try:
    rat = bsObj.find(name="span",attrs = {'class':"rating-other-user-rating"}).text
    rat=re.sub("\\n","",rat)
    rat=re.sub("/10","",rat)
    ratings.append(float(rat))
  except:
    ratings.append(np.nan)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ["Reviews","Ratings"])
df["Reviews"] = reviews
df["Ratings"] = ratings

df.head(10)
```
