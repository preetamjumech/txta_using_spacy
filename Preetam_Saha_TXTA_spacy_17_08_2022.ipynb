```python
import spacy #much faster than python , bottomlevel it is c, c++, must faster, optimizer
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://timesofindia.indiatimes.com/business/india-business/akasa-air-to-add-one-aircraft-every-two-weeks/articleshow/93614917.cms"
```

```python
header = {"USer-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}
doc = requests.get(url, header)
```

```python
bsObj = BeautifulSoup(doc.text,'html.parser')
bsObj
```

```python
news = bsObj.find(name='div',attrs={'class':'fewcent-93614917'}).text
news
```

```python
!python -m spacy download en_core_web_lg #download large language model
```

```python
nlp = spacy.load('en_core_web_lg') #load this language model #spacy is entirely model-based
```

```python
texts = nlp(news) #supply the text in the nlp pipeline
```

```python
for token in texts[:20]:
  print(token.text, token.pos_)
```

```python
list(texts.noun_chunks)[:20] #more informative than a single token 
```

```python
for chunk in texts.noun_chunks:
  if len(chunk)>5:
    print(chunk)
```

```python
for chunk in texts.noun_chunks:
  if len(chunk)>2:
    print(chunk)
```

```python
list(texts.sents) #breaking the texts into sentences
```

```python
for ent in texts.ents:
  print(ent.text)
```

```python
for ent in texts.ents:
  print(ent.text,ent.label_)
```

```python
for ent in texts.ents: #finding out different entities
  if ent.label_ =='PERSON':
    print(ent.text,ent.label_)
```

```python
for token in texts: #finding out dependency relationships
  print(token.dep_)
```

```python
for token in texts:
  print(token.text, token.dep_)
```

```python
for token in texts:
  if token.ent_type_ == "PERSON":
    print(token.text, token.dep_, " ".join([i.text for i in token.subtree]) )
```

```python
for token in texts:
  if token.ent_type_ == "PERSON":
    print(token.text, " ".join([i.text for i in token.subtree]) )
```

```python
for token in texts:
  if token.ent_type_ == "PERSON":
    txt = " ".join(i.text for i in token.subtree)
    if txt != token.text:
      print(token.text, " ".join([i.text for i in token.subtree]))
```
