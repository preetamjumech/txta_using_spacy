```python
url ="https://timesofindia.indiatimes.com/business/india-business/dubai-government-finalizes-a-memorandum-of-understanding-with-the-haryana/articleshow/94713513.cms"
```

```python
import requests
```

```python
doc = requests.get(url,header)
```

```python
bsObj = BeautifulSoup(doc.text,'html.parser')
```

```python
bsObj
```

```python
news = bsObj.find(name = "div", attrs = {"class":"fewcent-94713513"}).text
```

```python
news
```

```python
!python -m spacy download en_core_web_lg
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
texts = nlp(news)
```

```python
for token in texts[:30]:
    print(token.text, token.pos)  
```

```python
for token in texts[:30]:
    print(token.text, token.pos_)  
```

```python
list(texts.noun_chunks)[:10]
```

```python
for chunk in texts.noun_chunks:
    if len(chunk)>5:
        print(chunk)
```

```python
list(texts.sents)
```

```python
for ent in texts.ents:
    print(ent)
```

```python
for ent in texts.ents:
    print(ent.text)
```

```python
for ent in texts.ents:
    print(ent.text, ent.label_)
```

```python
for ent in texts.ents:
    if ent.label_ == "ORG":
        print(ent.text)
```

```python
for ent in texts.ents:
    if ent.label_ == "PERSON":
        print(ent.text)
```

```python
for token in texts:
    print(token.text, token.pos_)
```

```python
for token in texts:
    print(token.text, token.dep_)
```

```python
for token in texts:
    if token.ent_type_ == "PERSON":
        print(token.text, token.dep_, " ".join(i.text for i in token.subtree))
```

```python
for token in texts:
    if token.ent_type_ == "ORG":
        print(token.text, token.dep_, " ".join(i.text for i in token.subtree))
```

```python
for token in texts:
    if token.ent_type_ =="DATE":
        print(token.text, " ".join(i.text for i in token.subtree))
```

```python
for token in texts:
    if token.ent_type_ =="GPE":
        print(token.text, " ".join(i.text for i in token.subtree))
```

```python
for token in texts:
    if token.ent_type_ =="PERSON":
        txt = " ".join(i.text for i in token.subtree)
        if txt != token.text:
            print(token.text, " ".join(i.text for i in token.subtree))
        
```
